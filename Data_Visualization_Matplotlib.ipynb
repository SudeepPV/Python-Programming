# **Simple Line Plots**

Visualization of a single function y=f(x)

Two different ways of plotting the same sine curve in Matplotlib.

```python
fig = plt.figure()         # Create a figure
ax = plt.axes()            # Add a single subplot (Axes) to the figure
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))      # Plot sine curve on the Axes
plt.show()
```

* fig → the overall container for plots.

* ax → the actual plotting area (can have many inside one figure).

* Good for multiple plots and more control.

```python
#Using plt.plot() (State-Based Style)
plt.plot(x, np.sin(x))     # Directly plots to the current Axes
plt.show()
```

* Simpler for quick plots.

**Multiple Functions in One Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x))        # solid sine curve
plt.plot(x, np.cos(x), '--')  # dashed cosine curve
plt.show()
```

* Shows how to plot multiple lines in the same figure.

* Second argument '--' changes the line style.

#### Colors in Matplotlib

```python
fig = plt.figure()
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x - 0), '--', color='blue')       # color by name
plt.plot(x, np.sin(x - 1), color='g')                # short RGB code
plt.plot(x, np.sin(x - 2), color='0.75')             # grayscale (0=black, 1=white)
plt.plot(x, np.sin(x - 3), color='#FFDD44')          # hex color
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))    # RGB tuple
plt.plot(x, np.sin(x - 5), color='chartreuse')       # HTML color name
plt.show()
```

* Matplotlib supports named colors, short codes, grayscale, hex, RGB tuples, and HTML colors.

#### Line Styles in Matplotlib

```python
fig = plt.figure()
x = np.linspace(0, 10, 1000)

plt.plot(x, x + 0, linestyle='solid')    # long form
plt.plot(x, x + 1, linestyle='dashed')   # '--'
plt.plot(x, x + 2, linestyle='dashdot')  # '-.'
plt.plot(x, x + 3, linestyle='dotted')   # ':'

# Short codes for the same:
plt.plot(x, x + 4, '-')   # solid
plt.plot(x, x + 5, '--')  # dashed
plt.plot(x, x + 6, '-.')  # dashdot
plt.plot(x, x + 7, ':')   # dotted

plt.show()
```

* You can use either full names ('solid', 'dashed') or short codes ('-', '--').

Bubble scatter plot of the Iris dataset

* x-axis → sepal length (features[0])

* y-axis → sepal width (features[1])

* bubble size (s) → proportional to petal width (100 * features[3])

* bubble color (c) → represents the class (iris.target → 0=setosa, 1=versicolor, 2=virginica)

* colormap → 'viridis' (nice gradient from purple→yellow)

* alpha=0.2 → semi-transparency for overlapping points

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load iris dataset
iris = load_iris()
features = iris.data.T  # transpose so that each row is a feature

#iris.data.T transposes the data matrix so that each features[i] is an array of all values for the i-th feature.
#features[0] = sepal length
#features[1] = sepal width
#features[2] = petal length
#features[3] = petal width

# Bubble scatter plot
plt.scatter(
    features[0],                # sepal length
    features[1],                # sepal width
    alpha=0.2,                   # transparency
    s=100 * features[3],         # bubble size (scaled petal width)
    c=iris.target,               # color by species
    cmap='viridis'               # colormap
)

plt.xlabel(iris.feature_names[0])  # 'sepal length (cm)'
plt.ylabel(iris.feature_names[1])  # 'sepal width (cm)'
plt.colorbar(label='Species')      # optional: add color scale
plt.show()
```

This gives you an informative multivariate plot:

* You can visually see how species separate based on sepal length & width.

* The size of each point adds a third variable (petal width).

* The color encodes the species.

As datasets get larger than a few thousand points, plt.plot can be noticeably more efficient than plt.scatter. The reason is that plt.scatter has the capability to render a different size and/or color for each point, so the renderer must do the extra work of constructing each point individually. In plt.plot, on the other hand, the points are always essentially clones of each other, so the work of determining the appearance of the points is done only once for the entire set of data. For large datasets, the difference between these two can lead to vastly different performance, and for this reason, plt.plot should be preferred over plt.scatter for large datasets.



# **Density and Contour Plots**

The function is defined as:

$$
f(x, y) = \sin^{10}(x) + \cos(10 + y \cdot x) \cdot \cos(x)
$$

where:

- (x, y) are input variables,
- $sin^{10}(x)$ means $\left(\sin(x)\right)^{10}$.



```python
import numpy as np

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
#contour or surface plot

import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

# Create grid
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

# Plot
plt.contour(X, Y, Z, colors='black');
plt.colorbar()
plt.title('Contour plot of f(x,y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
plt.colorbar()
plt.title('Contour plot of f(x,y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of f(x,y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Our plot is looking nicer, but the spaces between the lines may be a bit distracting. We can change this by switching to a filled contour plot using the plt.contourf() function (notice the f at the end), which uses largely the same syntax as plt.contour().
Additionally, we'll add a plt.colorbar() command, which automatically creates an additional axis with labeled color information for the plot.



```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
plt.title('Contour plot of f(x,y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

One potential issue with this plot is that it is a bit "splotchy." That is, the color steps are discrete rather than continuous, which is not always what is desired. This could be remedied by setting the number of contours to a very high number, but this results in a rather inefficient plot

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis()
plt.title('Contour plot of f(x,y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

 It can sometimes be useful to combine contour plots and image plots. For example, here we'll use a partially transparent background image (with transparency set via the alpha parameter) and overplot contours with labels on the contours themselves.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

# Create grid points
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)

# Compute Z values
Z = f(X, Y)

# Plot contour lines
contours = plt.contour(X, Y, Z, levels=3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

# Overlay image plot with transparency
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', alpha=0.5)

# Add colorbar
plt.colorbar()

plt.title("Contour and Image plot of f(x, y)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# **Histograms, Binnings, and Density**

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # generate 1000 samples from standard normal

plt.hist(data)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(
    data,
    bins=30,               # use 30 bins
    density=True,          # normalize so area = 1 (probability density)
    alpha=0.5,             # semi-transparent bars
    histtype='stepfilled', # filled step plot style
    color='steelblue',     # bar fill color
    edgecolor='none'       # no edge color on bars
)
plt.show()
```

Parameters:

* bins=30: Divide data range into 30 bins.

* density=True: Plot probability density instead of counts.

* alpha=0.5: Transparency of bars (0=transparent, 1=opaque).

* histtype='stepfilled': Bars with a filled step outline (looks nicer).

* color='steelblue': Sets fill color.

* edgecolor='none': Removes borders around bars.

**Two-Dimensional Histograms and Binnings**

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the bivariate normal distribution
mean = [0, 0]
cov = [[1, 1], [1, 2]]

# Generate 10,000 samples from the distribution
x, y = np.random.multivariate_normal(mean, cov, 10000).T

# Plot 2D histogram with 30 bins and blue color map
plt.hist2d(x, y, bins=30, cmap='Blues')

# Add colorbar with label
cb = plt.colorbar()
cb.set_label('counts in bin')

plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Histogram of Bivariate Normal Samples')
plt.show()

# Compute counts matrix using numpy (same bins as above)
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
print(counts)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Generate 2D Gaussian data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

# Create a figure with 2 subplots: hexbin and KDE
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot hexbin on the first subplot
hb = axs[0].hexbin(x, y, gridsize=30, cmap='Blues')
cb0 = fig.colorbar(hb, ax=axs[0])
cb0.set_label('count in bin')
axs[0].set_title('Hexbin plot')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# Compute KDE on a grid
data = np.vstack([x, y])
kde = gaussian_kde(data)

xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot KDE density on the second subplot
im = axs[1].imshow(Z.reshape(Xgrid.shape),
                   origin='lower',
                   aspect='auto',
                   extent=[-3.5, 3.5, -6, 6],
                   cmap='Blues')
cb1 = fig.colorbar(im, ax=axs[1])
cb1.set_label('density')
axs[1].set_title('Kernel Density Estimate')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.tight_layout()
plt.show()
```

* np.random.multivariate_normal(mean, cov, 10000) draws 10,000 points from a 2D Gaussian with the specified mean and covariance.

* plt.hist2d() creates a 2D histogram heatmap with bin counts shown by color intensity.

* plt.colorbar() shows the scale of counts per bin.

* np.histogram2d() returns the counts matrix and bin edges, which you print.

**Creating multi-axes histogram plot**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate correlated bivariate normal data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up figure and GridSpec layout
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)

# Main scatter plot occupies most of the grid (top-right 3x3)
main_ax = fig.add_subplot(grid[:-1, 1:])

# Marginal histogram for y on the left (vertical)
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)

# Marginal histogram for x on the bottom (horizontal)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# Scatter plot on main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)
main_ax.set_xlabel('X')
main_ax.set_ylabel('Y')

# Histogram of x on the bottom (vertical bars)
x_hist.hist(x, bins=40, histtype='stepfilled', orientation='vertical', color='gray')
x_hist.invert_yaxis()  # Flip y-axis so histogram aligns with scatter

# Histogram of y on the left (horizontal bars)
y_hist.hist(y, bins=40, histtype='stepfilled', orientation='horizontal', color='gray')
y_hist.invert_xaxis()  # Flip x-axis to align with scatter

plt.show()
```

* Large scatter plot showing joint distribution of x and y.

* Small histogram on the bottom showing distribution of x.

* Small histogram on the left showing distribution of y.

* Axes are linked so zooming or panning in main plot also updates histograms.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime

# Load data (ensure 'births.csv' is in your working directory)
#Download the data from https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv.
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')

# Calculate quartiles for filtering outliers
quartiles = np.percentile(births['births'], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])

# Filter out extreme birth counts outside 5 standard deviations
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')

# Convert 'day' column to integer for date construction
births['day'] = births['day'].astype(int)

# Create datetime index from year, month, and day columns
births.index = pd.to_datetime(
    10000 * births.year + 100 * births.month + births.day,
    format='%Y%m%d'
)

# Aggregate births by month and day (ignoring year)
births_by_date = births.pivot_table('births', [births.index.month, births.index.day])

# Convert index to datetime objects with a dummy year (e.g., 2012)
births_by_date.index = [datetime(2012, month, day) for (month, day) in births_by_date.index]

# Plot the aggregated daily births over a year
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax)
ax.set_title("Average Daily Births (Filtered)")
ax.set_xlabel("Date")
ax.set_ylabel("Number of Births")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
from datetime import datetime

# Load data
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')

# Remove outliers based on IQR
quartiles = np.percentile(births['births'], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')

# Construct datetime index
births['day'] = births['day'].astype(int)
births.index = pd.to_datetime(
    10000 * births.year + 100 * births.month + births.day,
    format='%Y%m%d'
)

# Pivot table of average births by month and day (ignore year)
births_by_date = births.pivot_table('births', [births.index.month, births.index.day])
births_by_date.index = [datetime(2012, month, day) for (month, day) in births_by_date.index]

# Plot setup
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax)

# Style for holiday annotations
style = dict(size=10, color='gray')

# Annotate US holidays on the plot
ax.text(datetime(2012, 1, 1), 3950, "New Year's Day", **style)
ax.text(datetime(2012, 7, 4), 4250, "Independence Day", ha='center', **style)
ax.text(datetime(2012, 9, 4), 4850, "Labor Day", ha='center', **style)
ax.text(datetime(2012, 10, 31), 4600, "Halloween", ha='right', **style)
ax.text(datetime(2012, 11, 25), 4450, "Thanksgiving", ha='center', **style)
ax.text(datetime(2012, 12, 25), 3850, "Christmas", ha='right', **style)

# Label axes and title
ax.set(title='USA births by day of year (1969–1988)', ylabel='average daily births')

# Format x-axis with month ticks and labels
ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())  # hide major tick labels
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%b'))  # show short month names at mid-month

plt.show()
```
